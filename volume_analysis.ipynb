```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
from datetime import datetime,timedelta
%matplotlib inline
```

```python
end =datetime.today().date()
start =datetime.today()-timedelta(days=55)
start=start.date()
```

```python
nifty_all= pd.read_excel('all_stocks.xlsx','Sheet1')
nifty=list(nifty_all['Symbol'])
nifty
```

```python
final_disc={}
```

```python
for stock in nifty:
    nifty_share=pd.DataFrame()
    point=stock+'.NS'
    nifty_share = data.DataReader(point,'yahoo',start,end)
    nifty_share['Percentage']=(nifty_share['Volume']/nifty_share['Volume'].sum())*100
    transit=list(nifty_share['Percentage'].tail())
    final_disc[stock]=transit
    print(point)
```

```python
final_disc
```

```python
#average over last five days
average={}
for value in final_disc.keys():
    transit=sum(final_disc[value])
    average[value]=round(transit,2)
sort_average= dict(sorted((value, key) for (key,value) in average.items()))
sort_average
```

```python
#yesterday data
yesterday={}
for value in final_disc.keys():
    transit=final_disc[value][-1]
    yesterday[value]=round(transit,2)
sort_yesterday= dict(sorted((value, key) for (key,value) in yesterday.items()))
sort_yesterday
```

```python
#percentage growth***%volume growth
growth={}
transit=0
for value in final_disc.keys():
    second=final_disc[value][-1]
    first=final_disc[value][-2]
    try:
        transit=((second/first)-1)*100
    except ZeroDivisionError:
        transit=0
    growth[value]=transit*sum(final_disc[value])
sort_growth= dict(sorted((value, key) for (key,value) in growth.items()))
sort_growth
```
